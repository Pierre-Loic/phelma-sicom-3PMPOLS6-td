## Filtrage par convolution

Les **filtres de convolution** permettent d'appliquer des **transformations** sur des images. On peut par exemple **détecter les contours** grâce aux **filtres de Sobel** ou les **filtres laplacien**.

💻 *Exécuter le code qui affiche l'effet d'un filtre de Sobel vertical sur l'image en niveaux de gris. Quels sont les élèments mis en avant grâce au filtre de Sobel vertical ?*

```python
kernel = np.array([
    [-1,0,1],
    [-2,0,2],
    [-1,0,1],
])
img_2 = signal.convolve2d(greyscale_img, kernel, boundary='symm', mode='same')
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
ax[0].set_title('Image d\'origine')
ax[0].imshow(greyscale_img, cmap=plt.cm.gray)
ax[1].set_title("Image filtre de Sobel vertical")
ax[1].imshow(255-abs(img_2), cmap=plt.cm.gray) # A modifier pour améliorer le contraste 
plt.show()
```

💻 *Modifier le code pour avoir l'effet d'un filtre de Sobel horizontal sur l'image en niveaux de gris*

```python
kernel = np.array([
    [-1,-2,-1],
    [0,0,0],
    [1,2,1],
]) # Noyau à modifier
img_2 = signal.convolve2d(greyscale_img, kernel, boundary='symm', mode='same')
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
ax[0].set_title('Image d\'origine')
ax[0].imshow(greyscale_img, cmap=plt.cm.gray)
ax[1].set_title("Image filtre de Sobel horizontal")
ax[1].imshow(255-abs(img_2), cmap=plt.cm.gray)
plt.show()
```

💻 *Modifier le code pour avoir l'effet d'un filtre laplacien sur l'image en niveaux de gris*

```python
kernel = np.array([
    [0,1,0],
    [1,-4,1],
    [0,1,0],
]) # Noyau à modifier
img_2 = signal.convolve2d(greyscale_img, kernel, boundary='symm', mode='same')
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
ax[0].set_title('Image d\'origine')
ax[0].imshow(greyscale_img, cmap=plt.cm.gray)
ax[1].set_title("Image filtre de laplacien")
ax[1].imshow(255-abs(img_2), cmap=plt.cm.gray)
plt.show()
```

💻 *Construiser un filtre pour détecter les lignes diagonales (comme le poteau horizontal qui tient la caténaire du tram) sur l'image en niveaux de gris*

## Détection d'objets

Dans cette dernière partie, on va utiliser la notion de **filtre de convolution** vue précédemment pour **détecter un détail** à l'intérieur d'une image. L'exemple est constitué d'une **image contenant 400 chiffres** : environ **200 chiffres "1"**, **200 chiffres "4"** et **un chiffre "8"** (tous ces paramètres sont fixés par défaut mais ils peuvent être modifiés dans les paramètres de la fonction Python). L'objectif est de **détecter ce chiffre "8"**.

### Création de l'image

💻 *Exécuter le code suivant. Où se situe le chiffre 8 ? Comment de temps avez-vous mis pour le détecter ?*

```python
def create_seven_segment_display(number=0, sigma=1, norm=True):
    image = np.zeros((28, 28)) # Image noir de fond
    # Position des segments sur l'image
    segments = {
        'a': [(max(0, round(random.gauss(2, sigma))), i) for i in range(7, 21)],
        'b': [(i, min(27, round(random.gauss(22, sigma)))) for i in range(3, 13)],
        'c': [(i, min(27, round(random.gauss(22, sigma)))) for i in range(15, 25)],
        'd': [(min(27, round(random.gauss(25, sigma))), i) for i in range(7, 21)],
        'e': [(i, max(0, round(random.gauss(5, sigma)))) for i in range(15, 25)],
        'f': [(i, max(0, round(random.gauss(5, sigma)))) for i in range(3, 13)],
        'g': [(max(0, min(27, round(random.gauss(13, sigma)))), i) for i in range(7, 21)],
    }
    # Liste des positions des segments pour les valeurs héxadécimales
    digits = [
        "abcdef", "bc", "abged", "abgcd", "fgbc",
        "afgcd", "afgcde", "fabc", "abcdefg", "abcdfg",
        "efabcg", "fedcg", "afed", "bcdeg", "afged",
        "afge",
    ]
    # Création du chiffre choisi dans le paramètre "number"
    for seg in [segments[dig] for dig in digits[number]]:
        for y, x in seg:
            if norm is True:
                image[y, x] = 1
            else:
                image[y, x] = 255
    return image

def create_image(nb_1=1, nb_2=2, nb_unique=6):
    image = np.zeros((560, 560))
    rnd_number = random.randint(0, 400)
    for i in range(20):
        for j in range(20):
            if 20*j+i==rnd_number:
                number = nb_unique
            else:
                number = random.choice([nb_1, nb_2])
            img = create_seven_segment_display(number=number, sigma=1.5, norm=True)
            image[i*28:(i+1)*28,j*28:(j+1)*28]=img
    plt.imshow(image, cmap="gray")
    plt.axis('off')
    plt.show()
    return image

image = create_image(nb_1=1, nb_2=4, nb_unique=8)
```

### Création du modèle d'objet à détecter

L'étape suivante est de **créer le filtre de convolution** qui va permettre de **détecter le chiffre recherché**.

💻 *Exécuter le code suivant et faire varier l'écart-type `sigma`*

```python
kernel = create_seven_segment_display(number=8, sigma=1.5)*20
plt.imshow(kernel, cmap="gray")
plt.axis('off')
plt.show()
```

### Filtrage pour détecter l'objet

💻 *Exécuter le code suivant pour effectuer la convolution. Où se situe le chiffre 8 ?*

```python
img_2 = signal.convolve2d(image, kernel, boundary='symm', mode='same')
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
ax[0].set_title('Image d\'origine')
ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].axis('off')
ax[1].set_title("Image filtre de convolution")
ax[1].imshow(img_2, cmap=plt.cm.gray)
ax[1].axis('off')
plt.show()
```

💻 *Exécuter le code suivant pour augmenter le contraste*

```python
def plot_img(ax, pos, img, title_img, title_chart, I):
    # Affichage de l'image
    ax[0,pos].imshow(img, cmap='gray', vmin=0, vmax=255)
    ax[0,pos].set_title(title_img)
    # Affichage de l'histogramme
    sns.histplot(img.flatten(), kde=True, color="grey", ax=ax[1,pos])
    ax[1,pos].set_xlim(0, 255)
    ax[1,pos].set_xlabel("Intensité lumineuse des pixels")
    ax[1,pos].set_ylabel("")
    ax[1,pos].set_title(title_chart)
    # Ajout de barres verticales pour indiquer les limites du slider
    ax[1,pos].axvline(I[0], color='red', linestyle='--')
    ax[1,pos].axvline(I[1], color='red', linestyle='--')

@widgets.interact(I=widgets.IntRangeSlider(value=[190, 220], min=0, max=255, description='Intensité :'))
def plot(I):
    # Normaliser min et max des pixels
    image_equalized = np.minimum(np.maximum(img_2, I[0]), I[1])
    # Normaliser en sens inverse sur 0-255
    image_equalized_2 = ((image_equalized-I[0])/(I[1]-I[0])*255).astype("uint8")
    # Affichage des images
    fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 6))
    plot_img(ax, 0, img_2, "Image d'origine", "Histogramme de l'image d'origine", I)
    plot_img(ax, 1, image_equalized, f"Plage histogramme entre {I[0]} et {I[1]}", f"Histogramme avec intensité entre {I[0]} et {I[1]}", I)
    plot_img(ax, 2, image_equalized_2, "Image ramenée sur une plage 0-255", "Histogramme avec intensité entre 0 et 255", I)
    plt.show()
```
