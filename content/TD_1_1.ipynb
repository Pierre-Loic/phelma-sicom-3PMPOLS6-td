```python
colors_red = [(0, 0, 0), (1, 0, 0)]
colors_green = [(0, 0, 0), (0, 1, 0)]
colors_blue = [(0, 0, 0), (0, 0, 1)]
cm_red = LinearSegmentedColormap.from_list("Custom", colors_red, N=256)
cm_green = LinearSegmentedColormap.from_list("Custom", colors_green, N=256)
cm_blue = LinearSegmentedColormap.from_list("Custom", colors_blue, N=256)
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 6))
ax[0].imshow(img[:,:,0], cmap=cm_red)
ax[1].imshow(img[:,:,1], cmap=cm_green)
ax[2].imshow(img[:,:,2], cmap=cm_blue)
plt.show()
```

### Passage d'une image en couleurs  une image en niveaux de gris
Les images en **niveaux de gris** ne contiennent qu'**un seul canal**. Il existe **diffrentes manires** de passer d'une image **couleur**  une image en **niveaux de gris**.

 *Excuter le code qui calcule le niveau de gris en effectuant la moyenne des 3 canaux de couleurs primaires*

```python
greyscale_img_1 = img[:,:,0]/3 + img[:,:,1]/3 + img[:,:,2]/3
greyscale_img_1 = greyscale_img_1.astype(np.uint8)
plt.figure(figsize=(10, 5))
plt.imshow(greyscale_img_1, cmap=plt.cm.gray)
plt.show()
```

 *Modifier le code pour calculer le niveau de gris suivant le coefficient de luminance de la rec 709 (https://fr.wikipedia.org/wiki/Rec._709)*

```python
greyscale_img = .2126*img[:,:,0] + .7152*img[:,:,1] + .0722*img[:,:,2]
greyscale_img = greyscale_img.astype(np.uint8)
plt.figure(figsize=(10, 5))
plt.imshow(greyscale_img, cmap=plt.cm.gray)
plt.show()
```

### Perception d'une image
La **perception** que ltre humain a des couleurs dune image dpend non seulement de la **couleur elle-mme** mais aussi des autres **couleurs qui lentourent**. Cette notion est bien illustre par la **loi du contraste simultan**. L'exemple suivant montre une **illusion d'optique** base sur le contraste.

 *Excuter le code qui affiche l'illusion d'optique. Modifier la couleur du rectangle central, l'orientation du gradient (horizontal, vertical ou diagonal) et la couleur du gradient. Dans quelle configuration l'illusion d'optique est-elle la plus forte ?*

```python
# Dimensions de l'image
largeur, hauteur = 600, 400

# Cration d'une image avec un dgrad de gris
x = np.linspace(0, 1, largeur)
y = np.linspace(0, 1, hauteur)
X, Y = np.meshgrid(x, y)
Z = X * 255 # Cration d'un dgrad de gris

# Convertir en un tableau 3D pour une image RGB
image = np.stack((Z, Z, Z), axis=2).astype(np.uint8)

# Dfinition des dimensions et de la position du rectangle
largeur_rect = 500
hauteur_rect = 50
x_rect = (largeur - largeur_rect) // 2
y_rect = (hauteur - hauteur_rect) // 2

# Couleur du rectangle (gris uniforme)
couleur_rect = [128, 128, 128]  # Gris moyen

# Dessiner le rectangle dans l'image
image[y_rect:y_rect+hauteur_rect, x_rect:x_rect+largeur_rect] = couleur_rect

# Affichage de l'image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

### galisation d'histogrammes

Dans le cours, on a vu la notion d'**galisation d'histogramme**. Dans cette partie, on va effectuer un autre type de **transformation** de l'histogramme de l'image : on va **modifier la plage** de valeurs d'intensit lumineuse de l'image en niveaux de gris.

 *Excuter le code et modifier grce au slider la plage de valeur d'intensit minimale et maximale de l'histogramme de l'image en niveaux de gris. Etudier la diffrence d'impact lorsque la plage est grande ou petite.*

```python
greyscale_img_2 = 0.2125*img_2[:,:,0] + 0.7154*img_2[:,:,1] + 0.0721*img_2[:,:,2]

def plot_img(ax, pos, img, title_img, title_chart, I):
    # Affichage de l'image
    ax[0,pos].imshow(img, cmap='gray', vmin=0, vmax=255)
    ax[0,pos].set_title(title_img)
    # Affichage de l'histogramme
    sns.histplot(img.flatten(), kde=True, color="grey", ax=ax[1,pos])
    ax[1,pos].set_xlim(0, 255)
    ax[1,pos].set_xlabel("Intensit lumineuse des pixels")
    ax[1,pos].set_ylabel("")
    ax[1,pos].set_title(title_chart)
    # Ajout de barres verticales pour indiquer les limites du slider
    ax[1,pos].axvline(I[0], color='red', linestyle='--')
    ax[1,pos].axvline(I[1], color='red', linestyle='--')

@widgets.interact(I=widgets.IntRangeSlider(value=[100, 150], min=0, max=255, description='Intensit :'))
def plot(I):
    # Normaliser min et max des pixels
    image_equalized = np.minimum(np.maximum(greyscale_img_2, I[0]), I[1])
    # Normaliser en sens inverse sur 0-255
    image_equalized_2 = ((image_equalized-I[0])/(I[1]-I[0])*255).astype("uint8")
    # Affichage des images
    fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 6))
    plot_img(ax, 0, greyscale_img_2, "Image d'origine", "Histogramme de l'image d'origine", I)
    plot_img(ax, 1, image_equalized, f"Plage histogramme entre {I[0]} et {I[1]}", f"Histogramme avec intensit entre {I[0]} et {I[1]}", I)
    plot_img(ax, 2, image_equalized_2, "Image ramene sur une plage 0-255", "Histogramme avec intensit entre 0 et 255", I)
    plt.show()
```
